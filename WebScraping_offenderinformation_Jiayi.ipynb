```python
import pandas as pd
pd.set_option('max_rows', None)
pd.set_option('max_columns', None)
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import time
import requests
```

```python
data_1 = pd.read_excel('2.xlsx')
```

```python
data_2 = data_1[30:]
data_2
```

```python
data_3 = []
```

```python
for x in data_2[0]:
    time.sleep(3)
    last_text = requests.get(url = x, headers = {'User-Agent': 'Mozilla/5.0'}).text
    last_soup = BeautifulSoup(last_text,'lxml')
    data_3.append(last_soup.find_all('p')[-5].text)
```

```python
data_3
```

```python
data_4 = pd.DataFrame(data_3)
data_4
```

By using the same way:

```python
data_5 = pd.DataFrame(data_3)
data_5
```

```python
data_6 = pd.DataFrame(data_3)
data_6
```

```python
data_7 = pd.concat([data_4,data_5,data_6],axis = 1,keys = ['education','prior occupation','prior record'])
data_7
```
