# The dataset mainly records the last sentence of the death row inmates in Texas from 1984 to 2020. 

This notebook records the procedure of webscraping.

```python
import pandas as pd
pd.set_option('max_rows', None)
pd.set_option('max_columns', None)
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import requests
import time
import re
```

```python
web = 'https://www.tdcj.texas.gov/death_row/dr_executed_offenders.html'
req = Request(web, headers={'User-Agent': 'Mozilla/5.0'})
downloaded_web = urlopen(req).read()
```

```python
data_0 = pd.read_html(downloaded_web)[0]
data_0.head(5)
```

```python
data_1 = pd.DataFrame(data_0)
```

```python
soup = BeautifulSoup(downloaded_web)
```

```python
findlink_laststatement = re.compile(r'<a href="(.*?)"')
```

https://www.tdcj.texas.gov/death_row/dr_info/runnelstravislast.html
https://www.tdcj.texas.gov/death_row/dr_executed_offenders.html

```python
soup.find_all('a')
```

```python
data_2 = []
for item in soup.find_all('tr'):
    item = str(item)
    offenders = re.findall(findlink_laststatement,item)
    print(offenders)
    data_2.append(offenders)
```

```python
data_3 = data_2[1:]
data_3
```

```python
data_4 = pd.DataFrame(data_3, columns = ['Offender Information','Last Statement'])
data_4
```

```python
data_5 = pd.concat([data_4,data_1],axis=1)
data_6 = pd.DataFrame(data_5)
```

```python
data_7 = data_6.drop(columns = ['Link','Link.1'])
data_7
```

```python
data_8 = data_7.to_excel('Death Row Information.xlsx')
```
